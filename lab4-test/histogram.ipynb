```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

obraz = Image.open('obraz11.jpg')

blue_channel = np.array(obraz)[:, :, 2]
blue_50_count = np.sum(blue_channel == 50)

plt.figure(figsize=(8, 6))
plt.hist(blue_channel.ravel(), bins=256, color='blue')
plt.title("Histogram niebieskiego kanału")
plt.xlabel("Intensywność koloru")
plt.ylabel("Liczba pikseli")

plt.savefig('hist.png')
plt.show()
```
